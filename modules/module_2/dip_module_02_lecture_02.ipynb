## Statistics in Python

```python
data = np.array([1,2,3,4,5])
data.mean()
```

```python
N = len(data)
sums = data.sum()
x_bar = (sums)/N

np.allclose(data.mean(), x_bar)
```

```python
vect_mean = (data @ np.ones(N)) / N ## very pedantic way in linalg
np.allclose(data.mean(), vect_mean)
```

```python
np.median(data)
```

### Sturge's Rule
$$N_{bins} = 1 + \lceil log_2(n) \rceil$$

### Freedman-Daconis
$$W_{bin} = 2 (Q_{75} - Q_{25}) (n)^{-\frac{1}{3}}$$
$$N_{bins} = \frac{max(x) - min(x)}{W_{bin}}$$

```python
from scipy.fft import fft, fftfreq
import numpy as np
# Number of sample points
N = 600
# sample spacing
T = 1.0 / 500.0 
x = np.linspace(0.0, N*T, N, endpoint=False)
# y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x) + 2*np.sin(20.0 * 2.0*np.pi*x)
y = np.random.normal(x)
plt.plot(x,y)
```

```python
yf = fft(y) ## Fast Fourier Transform
xf = fftfreq(N, T)[:N//2]
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```

```python
counts, bins = np.histogram(y, bins=int(N/10))
plt.hist(bins[:-1], bins, weights=counts)
plt.show()
```
